```
#import Libraries needed
import pandas as pd
import seaborn as sns

#split data module
from sklearn.model_selection import train_test_split
```

```
#define path for data
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ["MPG", "Cylinders", "Displacement", "Horsepower", "Weight", "Acceleration", "Model Year","Origin"]
dataset = pd.read_csv(url, names= column_names, comment ='\t',
                      sep = " ", skipinitialspace = True,na_values = '?')
```

```
dataset.head()
```

```
dataset.Origin.value_counts()
```

```
dataset.isna().sum()
```

```
dataset = dataset.dropna()
dataset = dataset.reset_index(drop=True)
```

```
y = dataset.pop("MPG")
```

```
len(dataset)
```

```
X_train, X_test, y_train, y_test = train_test_split(dataset, y, test_size = 0.2, random_state = 42)
```

```
sns.pairplot(X_train[["Cylinders", "Displacement", "Acceleration"]], diag_kind='kde')
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

clf = LinearRegression(normalize=True)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)
```

```
print(r2_score(y_test, y_predict)*100)
```
